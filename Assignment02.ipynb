# 2번 문제
경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.

```python
def div_path(s):
    lst = []
    l = s.split('/')
    directory = l[0]
    
    for i in range(1,len(l)):
        if i == len(l)-1:
            file = l[i]
        else :
            directory = directory +'/'+l[i]
    
    lst.append(directory)
    lst.append(file)
    
    return lst
    
test1 = div_path('/usr/local/bin/python')
test2 = div_path('/home/chulsoo/test.txt')

print(test1)
print(test2)
```

- 문자열을 split 함수를 이용하여 /을 구분으로 나누어 리스트에 저장한다. 리스트의 마지막 원소을 제외한 문자열의 합이 경로명이기에 마지막 원소를 제외한 나머지 원소(문자열들)를 /과 함께 합쳐 directory로 저장한다. 마지막 원소(문자열)를 file로 저장한다. 이를 lst에 넣어 return 한다.

# 3번 문제
두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오

```python
def list_union(lista, listb):
    lst = lista
    
    for x1 in listb :
        add_result = True
        for x2 in lista:
            if x1 == x2 :
                add_result = False
        if add_result == True :
            lst.append(x1)
            
    lst.sort()
    return lst

testcase1 = list_union([1, 2, 3], [1, 2, 4])
testcase2 = list_union([-10, -5, 0, -1], [100, 9, 0, 9])
testcase3 =list_union([0, 1, 2], [0, 1, 2])

print(testcase1)
print(testcase2)
print(testcase3)
```

- listb의 원소를 lista의 원소와 하나하나 비교하여 중복되는 원소가 있을 시에는 추가하지 않았고, 중복되지 않는다면 lista의 원소를 포함하고 있는 lst에 listb의 원소를 추가하였다.

- 들어가는 리스트가 집합으로 가정하기 때문에, 아래와 같은 하나의 리스트에 중복되는 원소가 있는 경우는 없다고 생각하며 코드를 작성함
ex)[1,3,3,4,5],[1,3] -> [1,3,3]

# 4번 문제
두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '교집합'을 반환하는 함수 list_intersection(lista, listb)와 '차집합'을 반환하는 함수 list_difference(lista, listb)를 작성하시오.

```python
def list_difference(lista, listb):
    lst = []
    
    for x1 in listb :
        for x2 in lista:
            if x1 == x2 :
                lst.append(x1)
    lst.sort()
    
    return lst

testcase1 = list_difference([1, 2, 3], [1, 2, 4])
testcase2 = list_difference([-10, -5, 0, -1], [100, 9, 0, 9])
testcase3 =list_difference([0, 1, 2], [0, 1, 2])

print(testcase1)
print(testcase2)
print(testcase3)
```

- listb의 원소를 lista의 원소와 하나하나 비교하여 서로 중복되는 원소만을 lst에 추가하였다.

- 들어가는 리스트가 집합으로 가정하기 때문에, 아래와 같은 하나의 리스트에 중복되는 원소가 있는 경우는 없다고 생각하며 코드를 작성함
ex)[1,3,3,4,5],[1,3] -> [1,3,3]

# 5번 문제
두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.

```python
def print_primenumber(a, b):
    primenums = []
    
    for i in range (a,b):
        result = 1
        if i != 1 :
            for j in range(2,i):
                if i % j == 0 : 
                    result = 0
                    break
            if result == 1:
                primenums.append(i)
        
    print("{} 이상 {} 이하 정수 중 소수 리스트 : {}".format(a,b,primenums))

print_primenumber(2,1)
print_primenumber(1,13)
```

 - 입력받은 두 양의 정수 a,b 사이에 반복문이 돌도록 한다. 이 i가 1과 자기자신만을 약수로 가지는 즉, j가 2~(i-1)사이의 숫자에서 나머지 연산이 0이 아닐 때만, 리스트에 추가한다. 소수만이 추가된 리스트를 정렬 후, 출력한다.

# [Incremental Project 문제]
웹 URL로 지정된 웹페이지를 문자열로 가져와 1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.

```python
def check_case2(sNum, s, source):
    sNum = sNum + len(s) #< s > 제거
    while source[sNum:sNum+len('/'+s)] != '/'+ s : #</s> 전까지 제거
        sNum += 1
    return sNum


import requests
req = requests.get('https://www.naver.com/')
source = req.text

characters = "!?@#." #쓸모없는 특수문자
sNum = 0 # 문자열의 검색 위치 저장
text = '' # 의미있는 문자 저장

while  sNum < len(source) :
    #case1
    if source[sNum] == "<" :
        sNum += 1 
        #case 2
        if source[sNum:sNum+len('script')] == 'script' : #script 태그 속 내용 제거
            sNum = check_case2(sNum,'script',source)
        if source[sNum : sNum+len('style')] == 'style' : #style 태그 속 내용 제거
            sNum = check_case2(sNum,'style',source)
                
        while source[sNum] != '>' :
            sNum += 1
        sNum += 1
    
    #case3
    elif source[sNum] in characters : #특수문자 제거
        sNum += 1
        
    else : #위의 case가 아닐 시 의미있는 문자열에 추가해줌.
        text += source[sNum]
        sNum += 1

lst = text.split() 
print(lst, len(lst), end = '\n')
```

의미있는 문자열만을 뽑아내기 위해서 전체적으로 문자열의 인덱스 성질을 이용하여 코드를 작성하였다.


고려해야 하는 사항은 아래와 같다.
- case1. < ~~ > 꺽은 괄호 내에 있는 내용 제거
- case2. <> ~~ </> 의 태그 속 내용 제거 ex) style,script
- case3. 의미없는 특수문자 제거

문자열의 인덱스(sNum)를 수행할 때마다 1씩 증가 시켜준다.

 - case 1에 대해서는 < 를 발견 시 >을 발견하기 전까지 문자를 저장하지 않음으로써 내용을 제거한다.
 - case 2에 대해서는 특정 태그 안에서는 문자열이 아닌 것도 존재하기에, 특정 태그(script, style)를 지정하여 그 사이의 값은 문자를 저장하지 않음으로써 내용을 제거한다. 이 밖에도 html에서 style과 script와 같이 작동하는 것이 있다면 check_case2 함수를 이용하여 추가하면 된다.
 - case 3에 대해서는 의미없는 특수문자를 제거해주기 위하여, 의미없는 특수문자열에 해당하는 문자가 있다면 인덱스 값을 넘겨 내용을 제거한다.
