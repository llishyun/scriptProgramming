### Q5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)



```python
#1
# for ~ in 리터럴
def addall(list) :
    sum = 0
    for i in list :
        sum += i
    return sum
    
print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
# 리스트 내포방식
def addall(list) :
    list = [i for i in list]
    return sum(list)

print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
# 2
# for ~ in 리터럴
def addallodd(list) :
    sum = 0
    for i in list:
        if i%2==1:
            sum += i
    return sum
        
print (addallodd([1,3,5,6,7]))
print (addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
# 리스트 내포방식

def addallodd(list) :
    list = [i for i in list if i%2==1]
    return sum(list)

print (addallodd([1,3,5,6,7]))
print (addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
def length (X):
    return len(str(X))

def isStrange(num):
    if num < 0 : #음인 정수는 False 한다.
        return False
    
    else :
        num_len = length(num)
        if num_len == 1: #자릿수가 1일 경우 True (0도 포함)
            return True
        
        elif num_len > 1 :#자릿수가 1이상일 경우 조건 1을 검사하고 재귀적인 함수 구성을 통해 조건 2를 검사한다.
            if num%num_len == 0 and isStrange(int(num/num_len)):
                return True
            else :
                return False

print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
```

- case1. < ~~ > 꺽은 괄호 내에 있는 내용 제거 (<>가 중첩되는 경우에도 제거되어야 한다)
- case2. <> ~~ </> 의 태그 속 내용 제거 ex) style,script
- case3. 의미없는 특수문자 제거

문자열의 인덱스(sNum)를 수행할 때마다 1씩 증가 시켜준다.

 - case 1에 대해서는 < 를 발견 시 >을 발견하기 전까지 문자를 저장하지 않음으로써 내용을 제거한다.
 - case 2에 대해서는 특정 태그 안에서는 문자열이 아닌 것도 존재하기에, 특정 태그(script, style)를 지정하여 그 사이의 값은 문자를 저장하지 않음으로써 내용을 제거한다. 이 밖에도 html에서 style과 script와 같이 작동하는 것이 있다면 check_case2 함수를 이용하여 추가하면 된다. 또한 <>의 내부에 예를 들어 내용이 <1<2>3>과 구성되어 있다면 >를 만나 3의 내용을 제거하지 못하는 경우를 고려하여 중첩됐을 경우도 고려한다.

```python
import string
print(string.punctuation)
```

```python
def check_case2(sNum, s, source):
    sNum = sNum + len(s) #< s > 제거
    while source[sNum:sNum+len('/'+s)] != '/'+ s : #</s> 전까지 제거
        sNum += 1
    return sNum


def search_engine1(source):
   
    characters = set(string.punctuation) #쓸모없는 특수문자(string.punctuation 활용)
    sNum = 0 # 문자열의 검색 위치 저장
    text = '' # 의미있는 문자 저장

    while  sNum < len(source) :
        #case1
        if source[sNum] == "<" :
            cnt = 0
            sNum += 1 
            #case 2
            if source[sNum:sNum+len('script')] == 'script' : #script 태그 속 내용 제거
                sNum = check_case2(sNum,'script',source)
            if source[sNum : sNum+len('style')] == 'style' : #style 태그 속 내용 제거
                sNum = check_case2(sNum,'style',source)
                
            while source[sNum] != '>' or cnt != 0 :
                if source[sNum]== '<' : #태그가 중첩되었을 때 제거
                    cnt += 1
                elif source[sNum]=='>':
                    cnt -= 1
                sNum += 1
            sNum += 1

        #case3
        elif source[sNum] in characters : #특수문자 제거
            sNum += 1

        else : #위의 case가 아닐 시 의미있는 문자열에 추가해줌.
            text += source[sNum]
            sNum += 1

    lst = text.split() 
    wordCount = {}

    for word in lst :
        wordCount[word]=wordCount.get(word,0)+1
        
    print(wordCount)
    print(lst, len(lst))
```
