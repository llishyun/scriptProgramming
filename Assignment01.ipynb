# 8번 문제

```python
n = int(input("정수를 입력하세요 :"))

result = True

for i in range(2,n+1):
    for j in range(2,i):
        if i % j== 0 :
            result = False
            break
        else :
            result = True
                
    if result:
        print(i)
    
```

- 입력된 정수를 1과 입력된 정수 사이의 수 중 나머지 연산 값이 0이 한번도 나오지 않을 시, 소수라고 취급하여 출력한다. 

# 9번 문제

```python
s = input( "string S를 입력하세요 : ")
scores = {'Kevin': 0, 'Stuart' : 0 }

for i in range(len(s)):
    if s[i] in " AEIOU": # 모음일 때는 Kevin에게 더한다.
        scores['Kevin'] += len(s) - i 
    else: # 자음일 때는 Stuart에게 더한다.
        scores['Stuart'] += len(s) - i

        
if scores['Kevin'] > scores['Stuart']:
    print("승자 : Kevin 점수 :",scores['Kevin'])
elif scores['Kevin'] == scores['Stuart'] :
    print("승자 : None")
else : 
    print("승자 : Stuart 점수 :",scores['Stuart'])
```

- for문을 통해 입력된 문자열의 하나의 문자가 모음일 때는 Kevin에게, 자음일 때는 Stuart에게 마땅한 점수를 주고, 점수를 비교하여 승자를 출력해낸다.

# 10번 문제

```python
def appendDelete(s, t, k):
    n = 0;
    a = 0;
 #최소의 연산을 구한 후, 나머지 연산은 empty일 때 채운다고 가정하여 최소의 연산값을 구함
    if len(s) > len(t):
        for j in range(len(t),len(s)):
            s = s[0:len(s)-1]
            n += 1
    
    for i in range (0,len(s)):#하나씩 삭제하는 연산
        if s[i] != t[i] :
            a = i 
            for j in range(i,len(s)):
                s = s[0:len(s)-1]
                n += 1
            break

    if s != t: # 같지 않을 시, 하나씩 추가하는 연산
        for i in range(a,len(t)):
            s = s + t[i]
            n += 1
    
    if n <= k:
        return 'Yes'
    else :
        return 'No'
    
    
print(appendDelete('hackerhappy', 'hackerrank',9))
print(appendDelete('aba', 'aba',7))
print(appendDelete('ashley', 'ash',2))
```

- 먼저 s가 더 길 경우, s와 t의 글자 길이를 맞춘다. s와 t가 맞지 않는 원소부터 s의 끝까지 삭제한다. t가 될 수 있도록 필요한 글자를 추가한다.
- 위 코드는 최소의 연산을 구한 후, empty string 에서 한 글자를 빼는 과정이나, 글자를 빼고 다시 더하는 등(실질적인 의미가 없는 과정)을 횟수에 더한다고 생각하여, 최소의 연산이 k보다 작을 시, Yes를 출력한다.
