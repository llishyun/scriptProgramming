### Q2. 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

- 다형성(polymorphism)이란 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 호출에 대해 각각 다르게 반응하도록 하는 기능이다. 이러한 다형성을 이용하면 코드의 양을 줄이고, 여러 객체 타입을 하나의 타입으로 관리가 가능하다.

```python
#다형성 예제

class Car:
    def __init__(self, name):
        self.name = name
    def inform(self):
        print(self.name)
        
class Hyundai(Car):
    def inform(self):
        print(self.name + "'s brand is Hyundai")
        
class BMW(Car):
    def inform(self):
        print(self.name + "'s brand is BMW")

for i in (Hyundai('GRANDEUR'), BMW('BMW X3')) :
    i.inform()
```

### Q3.다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

- 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.

    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step : Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.
- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.
- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.
- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

```python
class Counter :
    
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    
    def __str__(self):
        return "[Count(step:{0})]{1}".format(self.step, self.count)
    
    def incr(self):
        self.count = self.count + self.step
        
    def __call__(self):
        self.incr()
        
    def __add__(self,num):
        self.count += num
        return self
        
    def __sub__(self,num):
        self.count -= num
        return self
  
    def __lt__(self,num):
        return self.count < num
    
    def __gt__(self,num):
        return self.count > num
    
    def __eq__(self,num):
        return self.count == num
    
    def __ne__(self,num):
        return self.count != num
```

```python
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
```

```python
c.incr()
d.incr()
print(c)
print(d)
```

```python
c()
d()
print(c)
print(d)
```

```python
c = c + 5
d = d - 5
print(c)
print(d)
```

```python
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

### Q4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#### MySet 설명
1. __init__ () 생성자
    - __init__ 함수는 생성자로서 실행된다. list의 원소를 append(e)를 통해 self로 추가해준다. list의 원소가 추가된 self를 eliminate_duplicate함수로 전달한다.
    
    
2. __eilminate_duplicate()
    - 정적메소드로 인스턴스를 생성하지 않고도 클래스명으로 호출이 가능하다. 인자로 받은 l의 원소를 하나씩 s에 넣어준다. 만약 인자가 s에 없다면 넣어주는 방식을 통해 중복을 막아준다. l을 ㅣ[:]=[]을 통해 비워주고 중복이 제거된 s의 원소를 다시 l에 추가하여 저장한다.

    
3. __str__():  
    - print로 출력하는 문자열을 표기하기 위해 __str__ 호출된다.  result에 기본 출력문 MySet: {을 담고 for e in self:를 통해서 self 리스트에 담긴 문자를 넣고 마지막에 " ,"을 없애기위해 result[0:len(result)-2]의 슬라이싱 연산을 통해 제거하여 result에 저장한다. return을 통해서 result 값을 반환해준다.

### Q5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오.정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.

- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

       >>> u = s | t 
       >>> print(u) 
       MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

       >>> u = s & t 
       >>> print(u) 
       MySet: {2, 3} 
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

       >>> s = MySet([1, 2, 3]) 
       >>> t = MySet([3, 4, 5]) 
       >>> u = s - t 
       >>> print(u) 
       MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    # | 연산자 오버로딩
    # self의 원소를 s에 옮겨주고, for문과 if문을 통해ㅣ의 원소들 중 s에 있지 않은 원소를 append로 추가한다.
    # 합집합의 결과인 MySet(s)를 반환한다.
    def __or__(self, l):
        s = []
        for e in self: 
            s.append(e)
        for e in l:
            if e not in s:
                s.append(e)
        return MySet(s)
    
    # & 연산자 오버로딩
    # for문과 if문을 통해 self의 원소와 l의 원소가 동일할 때에만 append로 추가한다.
    # 교집합의 결과인 MySet(s)를 반환한다.
    def __and__(self, l):
        s = []
        for e in self:
            for i in l:
                if e == i:
                    s.append(e)
        return MySet(s)
    

    # - 연산자 오버로딩
    # for문과 if문을 통해 self의 원소와 l의 원소가 겹치지 않을 때에만 append로 추가한다.
    # 차집합의 결과인 MySet(s)를 반환한다.
    def __sub__(self, l):
        s = []
        for e in self:
            if e not in l:
                s.append(e)
        return MySet(s)


if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    u = s|t
    print (u)
    u = s&t
    print (u)
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s-t
    print (u)
```
