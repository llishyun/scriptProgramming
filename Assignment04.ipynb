2) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다

```python
with open('s.txt','r')as f:
    with open('s1.txt','w')as f1:
        str_list = f.readlines()
        str_list.sort()
        f1.writelines(str_list)
    
with open('s1.txt','r')as f1:
    print(f1.read())
```

3) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

```python
with open('s.txt','r')as f:
    with open('s2.txt','w')as f1:
        str_list = f.readlines()
        str_list.sort(key = lambda x:x[1])
        f1.writelines(str_list)
    
with open('s2.txt','r')as f1:
    print(f1.read())
```

4) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.

```python
with open('s.txt','r')as f:
    with open('s3.txt','w')as f1:
        str_list = f.read()
        list1 = str_list.split()
        list0 = [" ".join(list1[3*i:3*(i+1)])+'\n'for i in range(4)]
        f1.writelines(list0)
    
with open('s3.txt','r')as f1:
    print(f1.read())
```

Q3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오

```python
def frange(start = 0.0, stop = 0.0, step = 0.1):
    lst = []
    if start > stop :
        stop,start = start, stop
    r = start
    while(r<stop):
        lst.append(round(r,2))
        r += step
    return lst

print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
```

Q4. 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

```python
def sum(*args) :
    sum = 0
    for i in args:
        sum +=i
        
    return sum

print(sum())
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1, 5, 7, 2, -10))
```

Q6. 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

```python
def myfact(n):
    if n == 1:      # n이 1일 때
        return 1    # 1을 반환하고 재귀호출을 끝냄
    return n * myfact(n - 1) 
```

```python
print(myfact(5))
print(myfact(10))
print(myfact(100))
```
